```python
def minmax(num1, num2):
    if num1 > num2:
        temp = num2
        num2 = num1
        num1 = temp
    
    for i in range(num1, num2 + 1):
        print(i)
```

```python
minmax(4,1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcplot(array1, array2):
    x = array1
    y = np.array([])
    for i in range(0, len(x)):
        y = np.append(y, np.cos(x[i]) * 5 * x[i])

    plt.plot(x, y)
    plt.show()
    
funcplot([1,2,3,4,5], [])
```
